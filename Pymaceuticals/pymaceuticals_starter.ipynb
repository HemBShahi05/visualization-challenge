## Observations and Insights 

* It can be observed from the regression plot between the weight of mouse and the tumour size of the mice treated with Capomulin drug that the tumour size has high corelation (0.84) with the weight of the mouse. This means that a mouse with a higher weight has a larger tumour.
* The results of the boxplot comparing the effectiveness of four different drugs (Capomulin, Ramicane, Infubinol and Ceftamin) shows that the drug Ramicane is most effective among these four drug types since the quartiles values for final tumour size is least for this drug. The drug Capomulin is also in a similar range and the final tumour size values are just slightly bigger than that for Ramicane drug. However, the other two drugs appear to be less effective.
* The data from the boxplot also shows that there are very less outliers in the data (only one for the drug Infubinol). 
* A lineplot with the volume of tumour vs number of days for a single mouse treated with Capomulin shows that the tumour size is decreasing almost linearly with increase in time. It shows the effectiveness of the drug against the tumour.


```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
%matplotlib inline

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
df_mouse = pd.merge(mouse_metadata, study_results, how='left', on=['Mouse ID', 'Mouse ID'] )

# Display the data table for preview
df_mouse.head()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumour volume 
# for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumour volume. 
# Assemble the resulting series into a single summary dataframe.

mean_tumor_size = clean_data.groupby(['Drug Regimen'])['Tumour Volume (mm3)'].mean()
median_tumor_size = clean_data.groupby(['Drug Regimen'])['Tumour Volume (mm3)'].median()
variance_tumor_size = clean_data.groupby(['Drug Regimen'])['Tumour Volume (mm3)'].var()
std_tumor_size = clean_data.groupby(['Drug Regimen'])['Tumour Volume (mm3)'].std()
sem_tumor_size = clean_data.groupby(['Drug Regimen'])['Tumour Volume (mm3)'].sem()

summary_tumour_size = pd.DataFrame({'Mean': mean_tumor_size,
                                    'Median': median_tumor_size,
                                    'Variance': variance_tumor_size,
                                    'Standard Deviation': std_tumor_size,
                                    'SEM': sem_tumor_size
                                   })
summary_tumour_size
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumour volume for each regimen
# Using the aggregation method, produce the same summary statistics in a single line

summary_tumour = clean_data.groupby(['Drug Regimen'])['Tumour Volume (mm3)'].agg(['mean','median','var','std','sem'])
summary_tumour
```

## Quartiles, Outliers and Boxplots

```python
clean_data.head()
```

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
# Start by getting the last (greatest) timepoint for each mouse
# Merge this group df with the original dataframe to get the tumour volume at the last timepoint

# Create a dataframe with the last timepoint for each mouse
last_timepoint = pd.DataFrame(clean_data.groupby(['Drug Regimen','Mouse ID'])[['Timepoint','Tumour Volume (mm3)']].last())
last_timepoint.rename(columns={'Timepoint':'Max Timepoint'}, inplace=True)
last_timepoint
```

```python
# reset the index of the datframe
last_timepoint = last_timepoint.reset_index()
last_timepoint
```

```python
# Put treatments into a list for for loop (and later for plot labels)
list_treatment_grp = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin' ]


# Create empty list to fill with tumour vol data (for plotting)
tumor_vol_capomulin = []
tumor_vol_ramicane = []
tumor_vol_infubinol = []
tumor_vol_ceftamin = []

for index, row in last_timepoint.iterrows():
    if row["Drug Regimen"] == 'Capomulin':
        tumor_vol_capomulin.append(row['Tumour Volume (mm3)'])
    if row["Drug Regimen"] == 'Ramicane':
        tumor_vol_ramicane.append(row['Tumour Volume (mm3)'])
    if row["Drug Regimen"] == 'Infubinol':
        tumor_vol_infubinol.append(row['Tumour Volume (mm3)'])
    if row["Drug Regimen"] == 'Ceftamin':
        tumor_vol_ceftamin.append(row['Tumour Volume (mm3)'])
        
```

```python
df_tumour_drugs = pd.DataFrame({'Capomulin':tumor_vol_capomulin,'Ramicane':tumor_vol_ramicane,
                              'Infubinol':tumor_vol_infubinol, 'Ceftamin' :tumor_vol_ceftamin})
df_tumour_drugs
```

```python
# Calculate the IQR and quantitatively determine if there are any potential outliers. 
# Write a function that gives all quartile values, iqr, upper and lower bounds and outliers as outputs

def quartiles(df):
    outlier = []
    q1, q2, q3 = df.quantile([.25, .5, .75])
    iqr = q3 -q1
    lower_bound = round(q1 - (1.5*iqr),2)
    upper_bound = round(q3 + (1.5*iqr),2)
    for x in df:
        if x < lower_bound or x > upper_bound:
            outlier.append(x)
        
    return f'First quartile (Q1): {q1:.2f}', f'Median (Q2): {q2:.2f}', f'third quartile (Q3): {q3:.2f}', f'IQR: {iqr:.2f}', f'Lower bound: {lower_bound:.2f}', f'Upper bound: {upper_bound:.2f}', f'Outlier: {outlier}'

quartiles(df_tumour_drugs['Capomulin'])
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumour volume for the Capomulin regimen
corr,_ = st.pearsonr(df_avg_tv['Weight (g)'], df_avg_tv['Tumour Volume (mm3)'])
print(f'The correlation coefficient between the mouse weight and average tumour volume is {corr:.3f}')
```

```python
# Linear regression model
x = df_avg_tv['Weight (g)']
y = df_avg_tv['Tumour Volume (mm3)']

res = st.linregress(x,y)
res
```

```python
# Regression plot
plt.scatter(x, y, color='navy',  label='actual data')
plt.plot(x, res.intercept + res.slope*x, 'green', label=f'y = {res.slope:.2f}x + {res.intercept:.2f} (fitted line)')
plt.xlabel('Weight of mouse in grams',fontsize = 12)
plt.ylabel('Volume of tumour in mm3',fontsize = 12)
plt.title('Regression plot of mice weight vs average tumour size', fontsize = 16, y=1.05)
plt.legend()
plt.tight_layout()
plt.show();
```
